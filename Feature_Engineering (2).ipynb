#1) What is a parameter?
A parameter is a variable in a model that the learning algorithm adjusts during training (e.g., weights in linear regression).

#2)What is correlation?
Correlation measures the strength and direction of the linear relationship between two variables.

Negative correlation: When one variable increases, the other decreases (e.g., speed ↑, travel time ↓).

#3)Define Machine Learning. What are the main components in Machine Learning?
Machine Learning is a technique where systems learn patterns from data without explicit programming.
Main components: Data, Model, Training algorithm, Loss function, and Evaluation.

#4)How does loss value help in determining whether the model is good or not?
Loss value measures the difference between predicted and actual output. A lower loss indicates a better model.

#5)What are continuous and categorical variables?

Continuous: Numerical values with infinite possibilities (e.g., height, salary).

Categorical: Values in categories or groups (e.g., gender, color).

#6)How do we handle categorical variables in Machine Learning?
By encoding them into numeric form.
Common techniques: Label Encoding, One-Hot Encoding, and Target Encoding.

#7)What do you mean by training and testing a dataset?

Training dataset: Used to train the model.

Testing dataset: Used to evaluate the model’s performance.

#8)What is sklearn.preprocessing?
It is a Scikit-learn module for data preprocessing tasks like encoding, scaling, normalization, and transformation.

#9)What is a Test set?
The test set is a portion of the dataset kept aside to evaluate the final model performance.

#10)How do we split data for model fitting (training and testing) in Python?
We use train_test_split from sklearn.model_selection. Example:

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


#11)How do you approach a Machine Learning problem?
Steps: Understand problem → Collect data → Preprocess → Split data → Train model → Evaluate → Improve.

#12)Why do we have to perform EDA before fitting a model to the data?
EDA (Exploratory Data Analysis) helps understand data distribution, detect missing values, outliers, and relationships between variables.

#13)What is correlation?
It is the statistical measure showing how two variables are related. Value ranges between -1 and +1.

#14)What does negative correlation mean?
It means when one variable increases, the other decreases. Example: Price ↑, demand ↓.

#15)How can you find correlation between variables in Python?
By using pandas.DataFrame.corr() or numpy.corrcoef() functions.

#16)What is causation? Explain difference between correlation and causation with an example.
Causation means one variable directly affects another.

Correlation: Ice cream sales ↑ and drowning cases ↑ (due to summer).

Causation: Rainfall ↑ → Crop growth ↑.

#17)What is an Optimizer? What are different types of optimizers?
An optimizer updates model parameters to minimize loss.
Examples:

Gradient Descent

Stochastic Gradient Descent (SGD)

Adam

RMSProp

#18)What is sklearn.linear_model?
It is a Scikit-learn module providing linear models like Linear Regression, Logistic Regression, Ridge, and Lasso.

#19)What does model.fit() do? What arguments must be given?
It trains the model using training data.
Arguments: Features (X) and Target (y).

Example: model.fit(X_train, y_train)

#20)What does model.predict() do? What arguments must be given?
It makes predictions on new data.
Argument: Input features (X).
Example: model.predict(X_test)

#21)What are continuous and categorical variables?

Continuous: Numerical (e.g., temperature).

Categorical: Labels or categories (e.g., city names).

#22)What is feature scaling? How does it help in Machine Learning?
Feature scaling normalizes feature ranges (e.g., 0–1 or standard scale). It helps models converge faster and improves accuracy.

#23)How do we perform scaling in Python?
Using StandardScaler or MinMaxScaler from sklearn. Example:

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


#24)What is sklearn.preprocessing?
It is used for preparing data, including encoding, scaling, normalization, and polynomial features.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


#26)Explain data encoding.
Data encoding transforms categorical values into numeric form so ML models can process them.
Examples: Label Encoding, One-Hot Encoding.
